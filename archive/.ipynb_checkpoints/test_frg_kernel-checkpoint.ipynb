# Test notebook for `frg_kernel.py`

This notebook builds a small mock patch basis and a mock spin-resolved patch vertex, then evaluates the three one-loop channel kernels:

- particle-particle (`pp`)
- particle-hole direct (`ph`)
- particle-hole crossed (`phc`)

The implementation follows the existing patch/channel conventions in `channels.py`, and uses a temperature-flow Matsubara approximation for the loop factor
\(L^\Lambda \sim T \sum_\omega [S^\Lambda G^\Lambda + G^\Lambda S^\Lambda]\).


```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np

sys.path.insert(0, str(Path('.').resolve()))

from patching import PatchPoint, PatchSet
from frg_kernel import (
    build_ph_channel_matrix,
    build_phc_channel_matrix,
    build_pp_channel_matrix,
    compute_ph_kernel,
    compute_phc_kernel,
    compute_pp_kernel,
)
```

```python
def make_mock_patchset(npatch: int = 12, radius: float = 1.0, energy_scale: float = 0.08):
    angles = np.linspace(0.0, 2.0 * np.pi, npatch, endpoint=False)
    patches = []
    for pid, ang in enumerate(angles):
        k = radius * np.array([np.cos(ang), np.sin(ang)], dtype=float)
        vf = np.array([-np.sin(ang), np.cos(ang)], dtype=float)
        eigvec = np.array([1.0, np.exp(1j * ang), np.exp(-1j * ang)], dtype=complex)
        eigvec = eigvec / np.linalg.norm(eigvec)
        patches.append(
            PatchPoint(
                patch_id=pid,
                k_cart=k,
                k_red=k.copy(),
                energy=energy_scale * np.cos(ang),
                vF=vf,
                vF_norm=float(np.linalg.norm(vf)),
                eigvec=eigvec,
                orbital_weight=np.abs(eigvec) ** 2,
            )
        )

    return PatchSet(
        mu=0.0,
        mu_used_for_contour=0.0,
        band_index=0,
        filling=0.5,
        patches=patches,
        fs_contour_k=np.array([p.k_cart for p in patches]),
        bz_vertices=np.zeros((6, 2), dtype=float),
        b1=np.array([2.0 * np.pi, 0.0], dtype=float),
        b2=np.array([0.0, 2.0 * np.pi], dtype=float),
        gauge_method='mock',
        gauge_loop_phase=0.0,
    )


patchsets = {
    'up': make_mock_patchset(),
    'dn': make_mock_patchset(),
}

N = patchsets['up'].Npatch
N
```

```python
rng = np.random.default_rng(7)

# Minimal set of spin blocks used in this notebook.
gamma = {}
for key in [
    ('up', 'up', 'up', 'up'),
    ('dn', 'dn', 'dn', 'dn'),
    ('up', 'dn', 'up', 'dn'),
    ('dn', 'up', 'dn', 'up'),
    ('up', 'dn', 'dn', 'up'),
    ('dn', 'up', 'up', 'dn'),
]:
    arr = rng.normal(scale=0.15, size=(N, N, N, N)) + 1j * rng.normal(scale=0.15, size=(N, N, N, N))
    gamma[key] = arr

Q = np.array([0.25, 0.10], dtype=float)
T = 0.08
Nw = 48
```

```python
pp_matrix = build_pp_channel_matrix(
    gamma,
    patchsets,
    Q,
    incoming_spins=('up', 'dn'),
    outgoing_spins=('up', 'dn'),
)
pp_kernel = compute_pp_kernel(
    gamma,
    patchsets,
    Q,
    incoming_spins=('up', 'dn'),
    outgoing_spins=('up', 'dn'),
    temperature=T,
    nfreq=Nw,
)

ph_matrix = build_ph_channel_matrix(
    gamma,
    patchsets,
    Q,
    incoming_spins=('up', 'up'),
    outgoing_spins=('up', 'up'),
)
ph_kernel = compute_ph_kernel(
    gamma,
    patchsets,
    Q,
    incoming_spins=('up', 'up'),
    outgoing_spins=('up', 'up'),
    temperature=T,
    nfreq=Nw,
)

phc_matrix = build_phc_channel_matrix(
    gamma,
    patchsets,
    Q,
    incoming_spins=('up', 'up'),
    outgoing_spins=('up', 'up'),
)
phc_kernel = compute_phc_kernel(
    gamma,
    patchsets,
    Q,
    incoming_spins=('up', 'up'),
    outgoing_spins=('up', 'up'),
    temperature=T,
    nfreq=Nw,
)
```

```python
def summarize_kernel(name, channel_matrix, kernel):
    print(f'[{name}]')
    print('  Q =', kernel.Q)
    print('  matrix shape =', kernel.matrix.shape)
    print('  channel-matrix norm =', np.linalg.norm(channel_matrix.matrix))
    print('  kernel norm =', np.linalg.norm(kernel.matrix))
    print('  max |kernel| =', np.max(np.abs(kernel.matrix)))
    print('  max patch residual =', np.max(kernel.residuals))
    if hasattr(kernel, 'hermitian_residual'):
        print('  hermitian residual =', kernel.hermitian_residual())
    print('  finite =', np.isfinite(kernel.matrix).all())
    print()

summarize_kernel('pp_ud', pp_matrix, pp_kernel)
summarize_kernel('ph_uu', ph_matrix, ph_kernel)
summarize_kernel('phc_uu', phc_matrix, phc_kernel)
```

```python
fig, axes = plt.subplots(3, 2, figsize=(10, 12))
entries = [
    ('pp channel matrix', np.abs(pp_matrix.matrix), pp_matrix.row_spins, pp_matrix.col_spins),
    ('pp one-loop kernel', np.abs(pp_kernel.matrix), pp_kernel.row_spins, pp_kernel.col_spins),
    ('ph channel matrix', np.abs(ph_matrix.matrix), ph_matrix.row_spins, ph_matrix.col_spins),
    ('ph one-loop kernel', np.abs(ph_kernel.matrix), ph_kernel.row_spins, ph_kernel.col_spins),
    ('phc channel matrix', np.abs(phc_matrix.matrix), phc_matrix.row_spins, phc_matrix.col_spins),
    ('phc one-loop kernel', np.abs(phc_kernel.matrix), phc_kernel.row_spins, phc_kernel.col_spins),
]

for ax, (title, mat, row_spins, col_spins) in zip(axes.ravel(), entries):
    im = ax.imshow(mat, origin='lower', aspect='auto')
    ax.set_title(title)
    ax.set_xlabel(f'col {col_spins}')
    ax.set_ylabel(f'row {row_spins}')
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
```

## Notes

- The notebook uses a **mock** vertex tensor, so the numerical values are not physical.
- What matters here is that the data path is correct: patch basis → channel matrix → loop contraction.
- For a real run inside your repo, replace `gamma` by the actual flowing patch-basis vertex tensor and `patchsets` by your real FS patch data.
