# fRG timing breakdown test notebook

This notebook is for **timing diagnosis only**.

Goal:
- test whether the main runtime cost is really in `compute_channel_rhs()` / one-loop kernel construction
- compare it against `diagnose_current_state()`
- further decompose the RHS cost into:
  - pp / phd / phc channel timing
  - per-`Q` timing
  - per-kernel timing for one selected `(channel, Q, spin block)`

This notebook uses your **current source files directly**.


```python
import time
import importlib
from collections import defaultdict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import noninteracting
import patching
import interaction
import kagome_order_diagnosis
import frg_flow
import frg_kernel

importlib.reload(noninteracting)
importlib.reload(patching)
importlib.reload(interaction)
importlib.reload(kagome_order_diagnosis)
importlib.reload(frg_kernel)
importlib.reload(frg_flow)

from noninteracting import KagomeNagaosa
from patching import FSPatcher, plot_patchset
from interaction import BareExtendedHubbard
from kagome_order_diagnosis import KagomeOrderDiagnoser
from frg_flow import FRGFlowSolver, BareVertexFromInteraction
from frg_kernel import FlowConfig, compute_pp_kernel, compute_ph_kernel, compute_phc_kernel
```

## 1. Build a standard test setup

This cell uses the same standard kagome setup, but with parameters chosen for timing diagnosis rather than final physics quality.

```python
# Standard kagome = KagomeNagaosa(phi=0)
model = KagomeNagaosa(parameters={"t": 1.0, "phi": 0.0}, spin=True, B=None)
filling = 5.0 / 12.0
mu = model.EF_from_filling(filling)

# Keep these modest so the timing notebook can finish.
Npatch = 12
grid_size = 220
U = 8.0
V = 0.2
nfreq = 8

t0 = time.perf_counter()
patcher_up = FSPatcher(
    model,
    band_index=1,
    filling=filling,
    orbital_slice=slice(0, 3),
    Npatch=Npatch,
    grid_size=grid_size,
    gauge_fix="parallel_transport",
    close_loop_gauge=True,
    verbose=False,
)
patcher_dn = FSPatcher(
    model,
    band_index=1,
    filling=filling,
    orbital_slice=slice(3, 6),
    Npatch=Npatch,
    grid_size=grid_size,
    gauge_fix="parallel_transport",
    close_loop_gauge=True,
    verbose=False,
)
patchset_up = patcher_up.build()
patchset_dn = patcher_dn.build()
patchsets = {"up": patchset_up, "dn": patchset_dn}
patch_time = time.perf_counter() - t0

bare_int = BareExtendedHubbard.from_kagome_model(model, U=U, V=V)
gamma_bare = BareVertexFromInteraction(bare_int, patchsets)
diagnoser = KagomeOrderDiagnoser(patchsets_by_spin=patchsets)

solver = FRGFlowSolver(
    patchsets=patchsets,
    bare_gamma=gamma_bare,
    diagnoser=diagnoser,
    T_start=5.0,
    T_stop=0.5,
    n_steps=8,
    temperature_grid="log",
    nfreq=nfreq,
    include_explicit_T_prefactor=True,
    diagnose_every=1,
    max_relative_update=1.0,
    min_substep_fraction=1.0/1000000.0,
    channel_divergence_threshold=50.0,
    eigenvalue_threshold=15.0,
)

print(f"patch build time = {patch_time:.3f} s")
print("Npatch =", Npatch)
print("nfreq =", nfreq)
print("n_pp_Q =", len(solver.pp_grid.q_list))
print("n_ph_Q =", len(solver.phd_grid.q_list))
print("n_phc_Q =", len(solver.phc_grid.q_list))
print("n_spin_blocks =", len(solver.spin_blocks))
print("spin_blocks =", solver.spin_blocks)
```

## 2. Time `compute_channel_rhs()` versus `diagnose_current_state()`

```python
T0 = float(solver.state.T)

t_rhs0 = time.perf_counter()
rhs_pp, rhs_phd, rhs_phc = solver.compute_channel_rhs(T0)
rhs_time = time.perf_counter() - t_rhs0

t_diag0 = time.perf_counter()
diag_payload = solver.diagnose_current_state()
diag_time = time.perf_counter() - t_diag0

compare_df = pd.DataFrame([
    {"component": "compute_channel_rhs", "time_sec": rhs_time},
    {"component": "diagnose_current_state", "time_sec": diag_time},
    {"component": "rhs/diag ratio", "time_sec": rhs_time / max(diag_time, 1e-12)},
])

display(compare_df)
print("diagnosis payload keys:", list(diag_payload.keys()))
```

## 3. Manually decompose RHS timing into pp / phd / phc

```python
cfg = solver._flow_config(T0)
gamma = solver.gamma_accessor()

rows = []

# pp
t0 = time.perf_counter()
rhs_pp_manual = solver._empty_channel_store(solver.pp_grid)
for iq, Q in enumerate(solver.pp_grid.q_list):
    for s1, s2, s3, s4 in solver.spin_blocks:
        ker = compute_pp_kernel(
            gamma,
            patchsets,
            Q,
            incoming_spins=(s1, s2),
            outgoing_spins=(s3, s4),
            config=cfg,
            allowed_spin_blocks=solver.allowed_spin_blocks,
        )
        rhs_pp_manual[(s1, s2, s3, s4, iq)] = np.asarray(ker.matrix, dtype=complex)
pp_time = time.perf_counter() - t0
rows.append({"channel": "pp", "time_sec": pp_time})

# phd
t0 = time.perf_counter()
rhs_phd_manual = solver._empty_channel_store(solver.phd_grid)
for iq, Q in enumerate(solver.phd_grid.q_list):
    for s1, s2, s3, s4 in solver.spin_blocks:
        ker = compute_ph_kernel(
            gamma,
            patchsets,
            Q,
            incoming_spins=(s1, s3),
            outgoing_spins=(s4, s2),
            config=cfg,
            allowed_spin_blocks=solver.allowed_spin_blocks,
        )
        rhs_phd_manual[(s1, s2, s3, s4, iq)] = np.asarray(ker.matrix, dtype=complex)
phd_time = time.perf_counter() - t0
rows.append({"channel": "phd", "time_sec": phd_time})

# phc
t0 = time.perf_counter()
rhs_phc_manual = solver._empty_channel_store(solver.phc_grid)
for iq, Q in enumerate(solver.phc_grid.q_list):
    for s1, s2, s3, s4 in solver.spin_blocks:
        ker = compute_phc_kernel(
            gamma,
            patchsets,
            Q,
            incoming_spins=(s1, s2),
            outgoing_spins=(s3, s4),
            config=cfg,
            allowed_spin_blocks=solver.allowed_spin_blocks,
        )
        rhs_phc_manual[(s1, s2, s3, s4, iq)] = np.asarray(ker.matrix, dtype=complex)
phc_time = time.perf_counter() - t0
rows.append({"channel": "phc", "time_sec": phc_time})

channel_time_df = pd.DataFrame(rows)
channel_time_df["fraction_of_rhs"] = channel_time_df["time_sec"] / channel_time_df["time_sec"].sum()
display(channel_time_df)
```

## 4. Time one selected kernel call for each channel

This isolates the cost of a single `(channel, Q, spin block)` build.

```python
spin_block = solver.spin_blocks[0]
Q_pp = solver.pp_grid.q_list[0]
Q_ph = solver.phd_grid.q_list[0]
Q_phc = solver.phc_grid.q_list[0]

s1, s2, s3, s4 = spin_block

rows = []

t0 = time.perf_counter()
ker_pp = compute_pp_kernel(
    gamma, patchsets, Q_pp,
    incoming_spins=(s1, s2),
    outgoing_spins=(s3, s4),
    config=cfg,
    allowed_spin_blocks=solver.allowed_spin_blocks,
)
rows.append({
    "channel": "pp",
    "Q_index": 0,
    "spin_block": str(spin_block),
    "time_sec": time.perf_counter() - t0,
    "max_abs": float(np.max(np.abs(ker_pp.matrix))),
})

t0 = time.perf_counter()
ker_phd = compute_ph_kernel(
    gamma, patchsets, Q_ph,
    incoming_spins=(s1, s3),
    outgoing_spins=(s4, s2),
    config=cfg,
    allowed_spin_blocks=solver.allowed_spin_blocks,
)
rows.append({
    "channel": "phd",
    "Q_index": 0,
    "spin_block": str(spin_block),
    "time_sec": time.perf_counter() - t0,
    "max_abs": float(np.max(np.abs(ker_phd.matrix))),
})

t0 = time.perf_counter()
ker_phc = compute_phc_kernel(
    gamma, patchsets, Q_phc,
    incoming_spins=(s1, s2),
    outgoing_spins=(s3, s4),
    config=cfg,
    allowed_spin_blocks=solver.allowed_spin_blocks,
)
rows.append({
    "channel": "phc",
    "Q_index": 0,
    "spin_block": str(spin_block),
    "time_sec": time.perf_counter() - t0,
    "max_abs": float(np.max(np.abs(ker_phc.matrix))),
})

single_kernel_df = pd.DataFrame(rows)
display(single_kernel_df)
```

## 5. Time per-`Q` cost inside each channel

This checks whether channel runtime is roughly proportional to the number of transfer momenta.

```python
per_q_rows = []

# pp per-Q
for iq, Q in enumerate(solver.pp_grid.q_list):
    t0 = time.perf_counter()
    for s1, s2, s3, s4 in solver.spin_blocks:
        _ = compute_pp_kernel(
            gamma, patchsets, Q,
            incoming_spins=(s1, s2),
            outgoing_spins=(s3, s4),
            config=cfg,
            allowed_spin_blocks=solver.allowed_spin_blocks,
        )
    per_q_rows.append({"channel": "pp", "iq": iq, "time_sec": time.perf_counter() - t0})

# phd per-Q
for iq, Q in enumerate(solver.phd_grid.q_list):
    t0 = time.perf_counter()
    for s1, s2, s3, s4 in solver.spin_blocks:
        _ = compute_ph_kernel(
            gamma, patchsets, Q,
            incoming_spins=(s1, s3),
            outgoing_spins=(s4, s2),
            config=cfg,
            allowed_spin_blocks=solver.allowed_spin_blocks,
        )
    per_q_rows.append({"channel": "phd", "iq": iq, "time_sec": time.perf_counter() - t0})

# phc per-Q
for iq, Q in enumerate(solver.phc_grid.q_list):
    t0 = time.perf_counter()
    for s1, s2, s3, s4 in solver.spin_blocks:
        _ = compute_phc_kernel(
            gamma, patchsets, Q,
            incoming_spins=(s1, s2),
            outgoing_spins=(s3, s4),
            config=cfg,
            allowed_spin_blocks=solver.allowed_spin_blocks,
        )
    per_q_rows.append({"channel": "phc", "iq": iq, "time_sec": time.perf_counter() - t0})

per_q_df = pd.DataFrame(per_q_rows)
display(per_q_df.groupby("channel")["time_sec"].agg(["count", "mean", "max", "min", "sum"]).reset_index())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

channel_order = ["pp", "phd", "phc"]
sum_df = channel_time_df.copy()
axes[0].bar(sum_df["channel"], sum_df["time_sec"])
axes[0].set_title("Total RHS time by channel")
axes[0].set_ylabel("time (sec)")

for ch in channel_order:
    sub = per_q_df[per_q_df["channel"] == ch]
    axes[1].plot(sub["iq"], sub["time_sec"], marker="o", label=ch)
axes[1].set_title("Per-Q time by channel")
axes[1].set_xlabel("Q index")
axes[1].set_ylabel("time (sec)")
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 6. Optional scaling tests

Run these only if you want to see whether timing follows the expected dependence on `Npatch` or `nfreq`.

```python
def build_test_solver(Npatch=8, nfreq=8, grid_size=180):
    model = KagomeNagaosa(parameters={"t": 1.0, "phi": 0.0}, spin=True, B=None)
    filling = 5.0 / 12.0

    pu = FSPatcher(
        model, band_index=1, filling=filling, orbital_slice=slice(0, 3),
        Npatch=Npatch, grid_size=grid_size,
        gauge_fix="parallel_transport", close_loop_gauge=True, verbose=False,
    )
    pdn = FSPatcher(
        model, band_index=1, filling=filling, orbital_slice=slice(3, 6),
        Npatch=Npatch, grid_size=grid_size,
        gauge_fix="parallel_transport", close_loop_gauge=True, verbose=False,
    )
    patchsets = {"up": pu.build(), "dn": pdn.build()}
    bare_int = BareExtendedHubbard.from_kagome_model(model, U=8.0, V=0.2)
    gamma_bare = BareVertexFromInteraction(bare_int, patchsets)
    diagnoser = KagomeOrderDiagnoser(patchsets_by_spin=patchsets)

    return FRGFlowSolver(
        patchsets=patchsets,
        bare_gamma=gamma_bare,
        diagnoser=diagnoser,
        T_start=5.0,
        T_stop=0.5,
        n_steps=8,
        temperature_grid="log",
        nfreq=nfreq,
        include_explicit_T_prefactor=True,
        diagnose_every=1,
        max_relative_update=1.0,
        min_substep_fraction=1.0/1000000.0,
        channel_divergence_threshold=50.0,
        eigenvalue_threshold=15.0,
    )

def time_rhs_once(sol):
    T0 = float(sol.state.T)
    t0 = time.perf_counter()
    _ = sol.compute_channel_rhs(T0)
    return time.perf_counter() - t0

def time_diag_once(sol):
    t0 = time.perf_counter()
    _ = sol.diagnose_current_state()
    return time.perf_counter() - t0
```

```python
# Scaling with Npatch (keep nfreq fixed)
npatch_list = [6, 8, 10]
rows = []

for npatch in npatch_list:
    sol = build_test_solver(Npatch=npatch, nfreq=8, grid_size=180)
    rhs_t = time_rhs_once(sol)
    diag_t = time_diag_once(sol)
    rows.append({
        "Npatch": npatch,
        "nfreq": 8,
        "rhs_time_sec": rhs_t,
        "diag_time_sec": diag_t,
        "n_pp_Q": len(sol.pp_grid.q_list),
        "n_ph_Q": len(sol.phd_grid.q_list),
        "n_phc_Q": len(sol.phc_grid.q_list),
    })

npatch_scaling_df = pd.DataFrame(rows)
display(npatch_scaling_df)
```

```python
# Scaling with nfreq (keep Npatch fixed)
nfreq_list = [4, 8, 12]
rows = []

for nf in nfreq_list:
    sol = build_test_solver(Npatch=8, nfreq=nf, grid_size=180)
    rhs_t = time_rhs_once(sol)
    diag_t = time_diag_once(sol)
    rows.append({
        "Npatch": 8,
        "nfreq": nf,
        "rhs_time_sec": rhs_t,
        "diag_time_sec": diag_t,
        "n_pp_Q": len(sol.pp_grid.q_list),
        "n_ph_Q": len(sol.phd_grid.q_list),
        "n_phc_Q": len(sol.phc_grid.q_list),
    })

nfreq_scaling_df = pd.DataFrame(rows)
display(nfreq_scaling_df)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(npatch_scaling_df["Npatch"], npatch_scaling_df["rhs_time_sec"], marker="o", label="rhs")
axes[0].plot(npatch_scaling_df["Npatch"], npatch_scaling_df["diag_time_sec"], marker="o", label="diag")
axes[0].set_xlabel("Npatch")
axes[0].set_ylabel("time (sec)")
axes[0].set_title("Scaling with Npatch")
axes[0].legend()

axes[1].plot(nfreq_scaling_df["nfreq"], nfreq_scaling_df["rhs_time_sec"], marker="o", label="rhs")
axes[1].plot(nfreq_scaling_df["nfreq"], nfreq_scaling_df["diag_time_sec"], marker="o", label="diag")
axes[1].set_xlabel("nfreq")
axes[1].set_ylabel("time (sec)")
axes[1].set_title("Scaling with nfreq")
axes[1].legend()

plt.tight_layout()
plt.show()
```

## How to interpret the timing results

You wanted to test the prediction:

> the dominant cost should be one-loop kernel construction / `compute_channel_rhs()`, not `order diagnosis`

This notebook is designed to test exactly that.

### If the prediction is correct, you should observe:
- `compute_channel_rhs()` slower than `diagnose_current_state()`
- channel timing roughly additive: `pp + phd + phc ≈ rhs total`
- per-`Q` timing within a given channel roughly similar
- RHS time grows strongly with `Npatch`
- RHS time also grows with `nfreq`

### If diagnosis is actually dominant, you will instead see:
- `diagnose_current_state()` comparable to or larger than RHS
- strong diagnosis-time growth when `diagnosis_Qs` is large

So this notebook should let you decide based on measurements, not guesses.
