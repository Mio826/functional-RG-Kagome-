
# run_frg_flow

This notebook wires together:

- `noninteracting.py`
- `patching.py`
- `interaction.py`
- `frg_flow.py`
- `kagome_order_diagnosis.py`

and runs a **temperature-flow patch fRG** test on a kagome model.

## Important practical note

The current solver is a **research scaffold**, not yet a production-optimized flow engine.
For first tests, keep:

- `Npatch` small, e.g. 8–16
- `nfreq` modest, e.g. 64–128
- `n_steps` small, e.g. 10–20

Once interfaces and physics are stable, scale up carefully.


```python

import numpy as np
import matplotlib.pyplot as plt

from noninteracting import KagomeKaneMeleSOC, KagomeNagaosa, KagomeStaggerFlux
from patching import FSPatcher, plot_patchset
from interaction import BareExtendedHubbard
from frg_flow import FRGFlowSolver, build_initial_gamma_from_interaction
```


## 1. Build a spinful noninteracting model

Pick one model below.  
For a first debug run, start with a simple weak-coupling case.


```python

# Example 1: weak spinful Kane-Mele-like kagome model
model = KagomeKaneMeleSOC(
    parameters={"t": 1.0, "l1": 0.05, "l2": 0.00},
    spin=True,
    B=0.0,
)

# Alternative examples:
# model = KagomeNagaosa(parameters={"t": 1.0, "phi": 0.15}, spin=True, B=0.0)
# model = KagomeStaggerFlux(parameters={"t": 1.0, "phi": 0.20}, spin=True, B=0.0)

filling = 0.50
mu = model.EF_from_filling(filling)
mu
```

## 2. Build spin-resolved patch sets

```python

# We patch the spin-up 3x3 block and spin-down 3x3 block separately.
# Keep Npatch small at first because the flow solver is expensive.

Npatch = 10

patch_up = FSPatcher(
    model,
    band_index=1,
    mu=mu,
    Npatch=Npatch,
    orbital_slice=slice(0, 3),
    gauge_fix="parallel_transport",
    verbose=False,
).build()

patch_dn = FSPatcher(
    model,
    band_index=1,
    mu=mu,
    Npatch=Npatch,
    orbital_slice=slice(3, 6),
    gauge_fix="parallel_transport",
    verbose=False,
).build()

patchsets = {"up": patch_up, "dn": patch_dn}

fig, axes = plt.subplots(1, 2, figsize=(10, 4.5))
plot_patchset(patch_up, ax=axes[0], show_velocity=False)
axes[0].set_title("Spin up patching")
plot_patchset(patch_dn, ax=axes[1], show_velocity=False)
axes[1].set_title("Spin down patching")
plt.tight_layout()
```

## 3. Build the bare antisymmetrized vertex

```python

interaction = BareExtendedHubbard.from_kagome_model(model, U=1.0, V=0.3)

gamma0 = build_initial_gamma_from_interaction(interaction, patchsets)

print("Spin blocks in Gamma0:")
for key, block in gamma0.items():
    print(key, block.shape, np.max(np.abs(block)))
```


## 4. Run the temperature flow

Design choices used here:

- **temperature grid**: logarithmic
- **update scheme**: explicit Euler with adaptive sub-stepping
- **instability criterion**: stop if either
  - `max |Gamma|` exceeds `gamma_divergence_threshold`, or
  - leading diagnosed channel eigenvalue exceeds `eigenvalue_threshold`


```python

solver = FRGFlowSolver(
    patchsets=patchsets,
    initial_gamma=gamma0,
    interaction=interaction,
    T_start=0.50,
    T_stop=0.05,
    n_steps=12,
    temperature_grid="log",
    nfreq=64,
    max_relative_update=0.08,
    gamma_divergence_threshold=1e2,
    eigenvalue_threshold=20.0,
    diagnose_every=1,
)

history = solver.run()
len(history)
```

```python

for rec in history:
    print(
        f"step={rec.step_index:02d}  "
        f"T={rec.temperature:.5f}  "
        f"|Gamma|_max={rec.gamma_norm:.4e}  "
        f"|lambda|max={rec.leading_eigenvalue_abs}  "
        f"order={rec.leading_order_label}  "
        f"instability={rec.instability}"
    )
    if rec.instability_reason is not None:
        print("   reason:", rec.instability_reason)
```

## 5. Plot flow diagnostics

```python

Ts = np.array([rec.temperature for rec in history], dtype=float)
gamma_norms = np.array([rec.gamma_norm for rec in history], dtype=float)
lead_eval = np.array([
    np.nan if rec.leading_eigenvalue_abs is None else rec.leading_eigenvalue_abs
    for rec in history
], dtype=float)

plt.figure(figsize=(6, 4))
plt.plot(Ts, gamma_norms, marker="o")
plt.xscale("log")
plt.yscale("log")
plt.gca().invert_xaxis()
plt.xlabel("T")
plt.ylabel(r"$\max |\Gamma|$")
plt.title("Vertex growth during flow")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(Ts, lead_eval, marker="o")
plt.xscale("log")
plt.yscale("log")
plt.gca().invert_xaxis()
plt.xlabel("T")
plt.ylabel("leading channel eigenvalue magnitude")
plt.title("Leading instability scale")
plt.tight_layout()
plt.show()
```

## 6. Inspect the detected instability

```python

last = history[-1]
print("Detected order label:", last.leading_order_label)
print("Leading channel name :", last.leading_channel_name)
print("Leading Q            :", last.leading_Q)
print("Instability          :", last.instability)
print("Reason               :", last.instability_reason)

payload = last.diagnosis_payload
payload
```


## 7. Suggested first validation path

Before large scans, check these:

1. weak coupling should not explode immediately
2. decreasing `max_relative_update` should make the flow smoother
3. changing `Npatch`, `nfreq`, and `n_steps` should not qualitatively destroy the leading trend
4. instability should come from a physically plausible channel, not random oscillation
