# Benchmark 3 — temperature loop-kernel sanity checks

This notebook tests the new `loop_kernel.py` module **before** writing a full RG flow.

Goals:
1. verify the module imports and builds pp/ph loop bubbles,
2. check that partner-patch matching is sensible,
3. inspect temperature dependence of the loop weights,
4. verify that the loop object can be used as a patch-diagonal contraction kernel.

This benchmark uses a **mock circular patch set** so it can run even if the full Kagome pipeline is not imported.
After this passes, the next step is to connect the same loop object to the running vertex update.


```python
import sys
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

repo_root = Path(".").resolve()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from loop_kernel import TemperatureLoopKernel
```

## 1. Build a lightweight mock patch set

The real repo patch objects usually carry `k_cart`, reciprocal vectors `b1,b2`, and a Fermi-velocity estimate.  
Here we only mimic the minimum interface needed by `TemperatureLoopKernel`.


```python
from dataclasses import dataclass

@dataclass
class MockPatch:
    k_cart: np.ndarray
    eigvec: np.ndarray
    vF: np.ndarray

class MockPatchSet:
    def __init__(self, patches, b1, b2):
        self.patches = list(patches)
        self.b1 = np.asarray(b1, dtype=float)
        self.b2 = np.asarray(b2, dtype=float)
        self.Npatch = len(self.patches)

def make_circular_patchset(N=24, radius=1.0, vf_mag=2.0):
    patches = []
    thetas = np.linspace(0.0, 2.0 * np.pi, N, endpoint=False)
    for th in thetas:
        k = radius * np.array([np.cos(th), np.sin(th)], dtype=float)
        v = vf_mag * np.array([np.cos(th), np.sin(th)], dtype=float)
        u = np.array([1.0, np.exp(1j*th), np.exp(-1j*th)], dtype=complex)
        u = u / np.linalg.norm(u)
        patches.append(MockPatch(k_cart=k, eigvec=u, vF=v))
    b1 = np.array([8.0, 0.0], dtype=float)
    b2 = np.array([0.0, 8.0], dtype=float)
    return MockPatchSet(patches, b1, b2)

patchsets = {
    "up": make_circular_patchset(N=24, radius=1.0, vf_mag=2.0),
    "dn": make_circular_patchset(N=24, radius=1.0, vf_mag=2.0),
}
print("Npatch(up) =", patchsets["up"].Npatch)
```

## 2. Instantiate the loop-kernel builder

```python
lk = TemperatureLoopKernel(
    patchsets,
    radial_cutoff=3.0,
    n_radial=801,
    dT_fraction=1e-3,
)
Q0 = np.array([0.0, 0.0])
T0 = 0.20
```

## 3. Build pp / ph loop bubbles at a test temperature

```python
pp = lk.pp_loop(Q0, spin_pair=("up", "dn"), T=T0)
phd = lk.ph_direct_loop(Q0, spin_pair=("up", "up"), T=T0)
phc = lk.ph_crossed_loop(Q0, spin_pair=("up", "up"), T=T0)

for name, bubble in [("pp", pp), ("ph_direct", phd), ("ph_crossed", phc)]:
    print(name, bubble.summary())
```

## 4. Plot the diagonal loop weights

These are the patch-diagonal weights that will later appear between two channel kernels in the flow update.

```python
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(pp.weights, marker="o", label="pp")
ax.plot(phd.weights, marker="s", label="ph_direct")
ax.plot(phc.weights, marker="^", label="ph_crossed")
ax.set_xlabel("internal patch index")
ax.set_ylabel("dT bubble weight")
ax.set_title(f"Loop-bubble weights at T={T0}")
ax.legend()
plt.show()
```

## 5. Check partner-patch matching residuals

```python
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(pp.residuals, marker="o", label="pp partner residual")
ax.plot(phd.residuals, marker="s", label="ph_direct partner residual")
ax.set_xlabel("patch index")
ax.set_ylabel("nearest-patch mismatch")
ax.set_title("Partner-patch matching residuals")
ax.legend()
plt.show()

print("max residual pp =", float(np.max(pp.residuals)))
print("max residual ph_direct =", float(np.max(phd.residuals)))
```

## 6. Temperature dependence sanity check

We now evaluate the same loop at several temperatures.  
This checks that the temperature dependence is generated at the **loop level**, even though the initial bare vertex itself has no temperature dependence.


```python
Ts = np.array([0.40, 0.25, 0.18, 0.12, 0.08, 0.05], dtype=float)
pp_mean = []
ph_mean = []
pp_max = []
ph_max = []

for T in Ts:
    bpp = lk.pp_loop(Q0, spin_pair=("up", "dn"), T=float(T))
    bph = lk.ph_direct_loop(Q0, spin_pair=("up", "up"), T=float(T))
    pp_mean.append(np.mean(bpp.weights))
    ph_mean.append(np.mean(bph.weights))
    pp_max.append(np.max(np.abs(bpp.weights)))
    ph_max.append(np.max(np.abs(bph.weights)))

fig, ax = plt.subplots(figsize=(7,4))
ax.plot(Ts, pp_mean, marker="o", label="mean pp weight")
ax.plot(Ts, ph_mean, marker="s", label="mean ph weight")
ax.invert_xaxis()
ax.set_xlabel("T")
ax.set_ylabel("mean dT bubble weight")
ax.set_title("Temperature dependence of loop weights")
ax.legend()
plt.show()

fig, ax = plt.subplots(figsize=(7,4))
ax.plot(Ts, pp_max, marker="o", label="max |pp weight|")
ax.plot(Ts, ph_max, marker="s", label="max |ph weight|")
ax.invert_xaxis()
ax.set_xlabel("T")
ax.set_ylabel("max absolute weight")
ax.set_title("Largest loop-weight magnitude vs temperature")
ax.legend()
plt.show()

print("Ts =", Ts.tolist())
print("pp_mean =", [float(x) for x in pp_mean])
print("ph_mean =", [float(x) for x in ph_mean])
```

## 7. Check matrix form

The returned object is diagonal in patch space.  
This is the object that will later be inserted between channel kernels in the RG update, schematically

$$\partial_T \Gamma \sim K_{\text{left}}\, \mathrm{diag}(w)\, K_{\text{right}}.$$

```python
Dpp = pp.as_diagonal_matrix()
Dph = phd.as_diagonal_matrix()

print("Diagonal matrix shapes:")
print("Dpp:", Dpp.shape)
print("Dph:", Dph.shape)

print("\nOff-diagonal norms (should be zero):")
print("||Dpp - diag(diag(Dpp))|| =", float(np.linalg.norm(Dpp - np.diag(np.diag(Dpp)))))
print("||Dph - diag(diag(Dph))|| =", float(np.linalg.norm(Dph - np.diag(np.diag(Dph)))))
```
