# Patching

```python
from noninteracting import KagomeKaneMeleSOC
from patching import FSPatcher, plot_patchset
import matplotlib.pyplot as plt

model = KagomeKaneMeleSOC(
    parameters={"t": 1.0, "l1": 0.2, "l2": 0.},
    spin=True,
    B=None,
)

filling = 5 / 12
band_index = 2

patcher = FSPatcher(
    model,
    band_index=band_index,
    filling=filling,
    grid_size=350,
    Npatch=48,
    fd_step=1e-4,
    orbital_slice=slice(0, 3),
    contour_min_points=40,
    contour_target_k=None,              # vanilla VHS 不强制围绕 Γ
    auto_level_shifts=[0.0, 1e-4, -1e-4, 5e-4, -5e-4],
    prefer_closed_contour=True,
    verbose=True,
)

patchset = patcher.build()

print("mu =", patchset.mu)
print("mu_used_for_contour =", patchset.mu_used_for_contour)
print("Npatch =", patchset.Npatch)
print("first patch k =", patchset.patches[0].k_cart)
print("first patch weight =", patchset.patches[0].orbital_weight)

fig, ax = plt.subplots(figsize=(6, 6))
plot_patchset(patchset, ax=ax, show_contour=True, show_velocity=False, show_bz=True)
plt.show()
```

```python
from noninteracting import KagomeStaggerFlux
from patching import FSPatcher, plot_patchset
import matplotlib.pyplot as plt

model = KagomeStaggerFlux(
    parameters={"t": 1.0, "phi": 0.4},
    spin=True,
    B=None,
)

filling = 5 / 12
band_index = 2

patcher = FSPatcher(
    model,
    band_index=band_index,
    filling=filling,
    grid_size=350,
    Npatch=48,
    fd_step=1e-4,
    orbital_slice=slice(0, 3),
    contour_min_points=40,
    contour_target_k=[0.0, 0.0],        # staggerflux 这里希望选围绕 Γ 的 pocket
    auto_level_shifts=[0.0, 1e-4, -1e-4, 5e-4, -5e-4],
    prefer_closed_contour=True,
    verbose=True,
)

patchset = patcher.build()

print("mu =", patchset.mu)
print("mu_used_for_contour =", patchset.mu_used_for_contour)
print("Npatch =", patchset.Npatch)
print("first patch k =", patchset.patches[0].k_cart)
print("first patch weight =", patchset.patches[0].orbital_weight)

fig, ax = plt.subplots(figsize=(6, 6))
plot_patchset(patchset, ax=ax, show_contour=True, show_velocity=False, show_bz=True)
plt.show()
```

```python
p = patchset.patches[0]

print("patch id =", p.patch_id)
print("k_cart =", p.k_cart)      # 笛卡尔动量
print("k_red  =", p.k_red)       # reduced coordinate
print("energy =", p.energy)      # 该 patch 点对应 band 的能量
print("vF     =", p.vF)
print("|vF|    =", p.vF_norm)
print("eigvec  =", p.eigvec)
print("weight  =", p.orbital_weight)
```
